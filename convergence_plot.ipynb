```python
from math import factorial, sin
import pandas as pd

sequence_1 = []
sequence_2 = []
sequence_3 = []
sequence_4 = []
sequence_5 = []
sequence_6 = []

for i in range(1, 60):
    sequence_1.append((3*i)**2/(i**2 + 1))
    sequence_2.append(2/factorial(i))
    sequence_3.append(3*(i)**2/(i**2 + 1))
    sequence_4.append((-1)**(i)*50/(i))
    sequence_5.append(70*i/(i + 2**(i/2)))
    sequence_6.append(40*sin(i)/i)


df_1 = {'p_1' : sequence_1, 'p_2' : sequence_3, 'p_3' : sequence_5}
df_2 = {'parameter_1' : sequence_1, 'parameter_2' : sequence_2, 'parameter_3' : sequence_3}
df_3 = {'param_1' : sequence_4, 'param_2' : sequence_5, 'param_3' : sequence_6}
a = pd.DataFrame(df_1)
b = pd.DataFrame(df_2)
c = pd.DataFrame(df_3)
```

```python
def convergence_plot(data_df):
    """Make a coonvergence plot.
    
    Args:
        data_df (DataFrame): The column names are the names of different methods and each 
        column contains the optimization values of each method from each iteration.
        
    """

    from bokeh.models import ColumnDataSource
    from bokeh.palettes import Viridis256
    from bokeh.plotting import figure, output_notebook, show
    import random

    #Make dictionary to link plot lines with ColumnDataSource
    number_columns = data_df.shape[1]
    columns = list(data_df.columns)
    colors = random.sample(Viridis256, number_columns)

    xs = [[] for i in range(number_columns)]
    ys = [[] for i in range(number_columns)]
    legends = []

    for i, column in enumerate(data_df.columns):
        ys[i] = list(data_df[column])
        xs[i] = list(data_df.index)
        legends.append(column)

    source_data = {'xs': xs, 'ys': ys, 'colors': colors, 'legends': legends}
    source = ColumnDataSource(source_data)
    
    #Make plot
    p = figure(plot_height=350,tools="tap,reset,save", x_range=[0, (len(data_df)+50)])
    p.title.text = 'Convergence Plot'

    p.multi_line(xs='xs', ys='ys', color='colors', line_width=4, legend='legends', 
                 nonselection_alpha=0, source=source)

    output_notebook()

    return show(p)
```

```python
convergence_plot_1 = convergence_plot(a)
```

```python
convergence_plot_2 = convergence_plot(b)
```

```python
convergence_plot_3 = convergence_plot(c)
```
